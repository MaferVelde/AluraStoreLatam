# **Conclusiones**

En base a nuestro análisi obtuvimos los siguientes datos:

*  Los ingresos totales de las tiendas son:


   1. Tienda #1: 1,150,880,400.00
   2. Tienda #2: 1,116,343,500.00
   3. Tienda #3: 1,098,019,600.00
   4. Tienda #4: 1,038,375,700.00


   Por lo tanto la tienda con mayores ingresos es la Tienda #1, la que tuvo menos ingresos es la Tienda #4.


*  Las categorías de productos más vendidas en general son la categoría de "Muebles" y la de "Electrónicos" en todas las tiendas, y las menos vendidas son la categoría "Artículos para el hogar" en las primeras 3 tiendas y la categoría de "Instrumentos musicales" en la Tienda #4.


*  Las calificaciones promedio de los clientes por tienda son:


   1. Tienda #1: 3.98
   2. Tienda #2: 4.04
   3. Tienda #3: 4.05
   4. Tienda #4: 4.00

   El ranking de calificación de las tiendas de mejor a peor calificación es: Tienda #3, Tienda #2, Tienda #4 y Tienda #1.
   
*  Los productos más vendidos son: "TV LED UHD 4K", "Iphone 15" y "Refrigerador"; y los menos vendidos fueron: "Cuerda para saltar" y "Cubo mágico 8x8".

*  El coste de envío promedio para cada tienda es:

   1. Tienda #1: 26018.61
   2. Tienda #2: 25216.24
   3. Tienda #3: 24805.68
   4. Tienda #4: 23459.46

*  En base a las coordenadas de mayor ventas podemos observar que en general la mayor cantidad de ventas para todas las tiendas se encuentra en el mismo sector, que se encuentra en su mayoria en la zona más cercana a "Bogotá" y de menor ventas en los sectores más alejados como por ejemplo "Valledupar"; al mismo tiempo si revisamos las calificaciones obtenidas en base al "Lugar de compras", la **Tienda 1** y **Tienda 3** son las que han obtenido mejor calificación en esos sectores, la **Tienda 2** y **Tienda 4** obtuvieron mejores calificaciones en los sectores con menores ventas.

En base a estos datos, podemos concluir que la tienda #4 es la que genera menos ganancias, además de ser de las tiendas con calificación más baja en general, sus mejores calificaciones son en sectores con menor cantidad de ventas, por ese motivo recomendamos que sea vendida la tienda #4.
